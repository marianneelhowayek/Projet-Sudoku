```python
import tkinter as tk

def creation_grille():
    case_taille = 50
    entries = []

    space = 5
    for row in range(9):
        y_space = row * case_taille + (row // 3) * space
        row_entries = []
        for col in range(9):
            x_space = col * case_taille + (col // 3) * space
            x = 2 + x_space + case_taille / 2
            y = 2 + y_space + case_taille / 2
            entry = tk.Entry(window, justify="center")
            entry.config(font=("Tekton Pro", 20), fg='black', bg="#f2f2f2")
            canva.create_window(x, y, window=entry, width=case_taille, height=case_taille, )
            row_entries.append(entry)
        entries.append(row_entries)

def format_timer(time: int):
    h = time // 3600
    m = (time % 3600) // 60
    s = (time % 3600) % 60
    return f"{h}:{m:0>2}:{s:0>2}"

def update_timer():
    global elapsed_time
    elapsed_time += 1
    timer_label.config(text=format_timer(elapsed_time))
    window.after(1000, update_timer)

def menu():
    reussite = tk.Canvas(window, width=120, height=150, bg='#f2f2f2', highlightthickness=0)
    reussite.place(relx=0.2, rely=0.25, anchor="center")

    quitter = tk.Button(reussite, text="Quitter", borderwidth= 0,
                        bg="#047274", fg= "#f2f2f2", activebackground= "#047274", activeforeground= "#f2f2f2")
    grille = tk.Button(reussite, text="Nouvelle grille", borderwidth= 0,
                       bg="#047274", fg= "#f2f2f2", activebackground= "#047274", activeforeground= "#f2f2f2")
    rejouer = tk.Button(reussite, text="Rejouer", borderwidth= 0,
                        bg="#047274", fg= "#f2f2f2", activebackground= "#047274", activeforeground= "#f2f2f2")
    resoudre = tk.Button(reussite, text="RÃ©soudre la grille", borderwidth= 0,
                         bg="#047274", fg= "#f2f2f2", activebackground= "#047274", activeforeground= "#f2f2f2")

    resoudre.place(relx=0.5, rely=0.2, anchor="center")
    rejouer.place(relx=0.5, rely=0.4, anchor="center")
    grille.place(relx=0.5, rely=0.6, anchor="center")
    quitter.place(relx=0.5, rely=0.8, anchor="center")

window = tk.Tk()
window.title("Sudoku")
window.resizable(False, False)

window.geometry("550x750")
window.minsize(480,360)
window.config(background='#047274')
window.iconbitmap("logo.ico")


canva = tk.Canvas(window, width=460, height=460, bg= "#047274")
canva.place(relx=0.5, rely=0.2, anchor="n")

elapsed_time = 0
timer_label = tk.Label(window, text="0:00:00", font=("Tekton Pro", 16), bg="#047274", fg="#f2f2f2")
timer_label.place(relx=0.5, rely=0.1, anchor="center")

mistakes_label = tk.Label(window, text="mistakes: 0", font=("Tekton Pro", 16), bg="#047274", fg="#f2f2f2")
mistakes_label.place(relx=0.5, rely=0.91, anchor="center")

eraser = tk.PhotoImage(file= "tools/eraser.png")
eraser_button = tk.Button(window, image= eraser, borderwidth= 0)
eraser_button.place(relx=0.84, rely=0.88)

hint = tk.PhotoImage(file= "tools/hint.png")
hint_button = tk.Button(window, image= hint, borderwidth= 0)
hint_button.place(relx=0.2, rely=0.07)

restart = tk.PhotoImage(file= "tools/restart.png")
restart_button = tk.Button(window, image= restart, borderwidth= 0)
restart_button.place(relx=0.28, rely=0.07)

settings = tk.PhotoImage(file= "tools/settings.png")
settings_button = tk.Button(window, image= settings, borderwidth= 0, command= menu)
settings_button.place(relx=0.12, rely=0.07)

update_timer()
creation_grille()

window.mainloop()
```
