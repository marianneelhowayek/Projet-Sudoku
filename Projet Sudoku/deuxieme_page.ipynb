```python
import tkinter as tk

def creation_grille():
    case_taille = 50
    sudoku_taille = 9 * case_taille

    for ligne in range(10): # 9 : nb de lignes
        if ligne % 3 == 0:
            epaisseur = 3
        else:
            epaisseur = 1
        grille.create_line(0, ligne * case_taille, sudoku_taille, ligne * case_taille, width= epaisseur)

    for col in range(10): # 9 : nb de colonne
        if col % 3 == 0:
            epaisseur = 3
        else:
            epaisseur = 1
        grille.create_line(col * case_taille, 0, col * case_taille, sudoku_taille, width= epaisseur)

def format_timer(time: int): 
    if time == "-":
        return "-"
    else:
        h = time // 3600
        m = (time % 3600) // 60
        s = (time % 3600) % 60
        return f"{h}:{m:0>2}:{s:0>2}"

def update_timer():
    global elapsed_time
    elapsed_time += 1
    timer_label.config(text=format_timer(elapsed_time))
    window.after(1000, update_timer)

window = tk.Tk()
window.title("Sudoku")

window.geometry("550x750")
window.minsize(480,360)
window.config(background='#047274')
window.iconbitmap("logo.ico")

grille = tk.Canvas(window, width=450, height=450, bg="#f2f2f2")
grille.place(relx=0.1, rely=0.2)

one = tk.PhotoImage(file= "numbers/one.png")
two = tk.PhotoImage(file= "numbers/two.png")
three = tk.PhotoImage(file= "numbers/three.png")
four = tk.PhotoImage(file= "numbers/four.png")
five = tk.PhotoImage(file= "numbers/five.png")
six = tk.PhotoImage(file= "numbers/six.png")
seven = tk.PhotoImage(file= "numbers/seven.png")
eight = tk.PhotoImage(file= "numbers/eight.png")
nine = tk.PhotoImage(file= "numbers/nine.png")

one_button = tk.Button(window, image= one, highlightthickness= 0)
two_button = tk.Button(window, image= two, highlightthickness= 0)
three_button = tk.Button(window, image= three, highlightthickness= 0)
four_button = tk.Button(window, image= four, highlightthickness= 0)
five_button = tk.Button(window, image= five, highlightthickness= 0)
six_button = tk.Button(window, image= six, highlightthickness= 0)
seven_button = tk.Button(window, image= seven, highlightthickness= 0)
eight_button = tk.Button(window, image= eight, highlightthickness= 0)
nine_button = tk.Button(window, image= nine, highlightthickness= 0)

one_button.place(relx=0.1, rely=0.88)
two_button.place(relx=0.19, rely=0.88)
three_button.place(relx=0.28, rely=0.88)
four_button.place(relx=0.37, rely=0.88)
five_button.place(relx=0.46, rely=0.88)
six_button.place(relx=0.55, rely=0.88)
seven_button.place(relx=0.64, rely=0.88)
eight_button.place(relx=0.73, rely=0.88)
nine_button.place(relx=0.82, rely=0.88)

# digit = {1 : one, 2 : two, 3 : three, 4 : four, 5 : five, 6 : six, 7 : seven, 8 : eight, 9 : nine}
# digit_button = {1 : one_button, 2 : two_button, 3 : three_button, 4 : four_button, 5 : five_button,
#                 6 : six_button, 7 : seven_button, 8 : eight_button, 9 : nine_button}

elapsed_time = 0
timer_label = tk.Label(window, text="0:00:00", font=("cranberry", 16), bg="#047274", fg="#f2f2f2")
timer_label.place(relx=0.5, rely=0.1, anchor= "center")

mistakes_label = tk.Label(window, text="mistakes :", font=("cranberry", 16), bg="#047274", fg="#f2f2f2")
mistakes_label.place(relx=0.8, rely=0.1, anchor= "center")

update_timer()

creation_grille()

window.mainloop()
```
